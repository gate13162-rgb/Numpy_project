```python
import pandas as pd
import numpy as np
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}
students_df = pd.DataFrame(students_data)

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}
enrollments_df = pd.DataFrame(enrollments_data)

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}
scores_df = pd.DataFrame(scores_data)

null_count=students_df.isnull().sum()
null_percentage=students_df.isnull().mean()*100
null_percentage=null_percentage.round(2)
print("Null Count:\n",null_count)
print("Null Percentage:\n",null_percentage)

students_df['city'] = students_df['city'].fillna('Unknown')
print (students_df['city'])

students_df = students_df.dropna(subset=['name'])
print (" The cleaned student dataframe :\n", students_df)

print("\n \n \tTASK 2: Multiple Join Operations")

merged_df = pd.merge(students_df,enrollments_df, on='student_id', how='inner')
print("\n\n\nMerged Inner DataFrame:\n", merged_df)

merged_left_df = pd.merge(students_df,enrollments_df, on='student_id', how='left')
print("\n\n\nMerged Left DataFrame:\n", merged_left_df)

merged_right_df = pd.merge(students_df,enrollments_df, on='student_id', how='right')
print("\n\n\nMerged Right DataFrame:\n", merged_right_df)

merged_fullouter_df = pd.merge(students_df,enrollments_df, on='student_id', how='outer',indicator= 'True')
print("\n\n\nMerged Full Outer DataFrame:\n", merged_fullouter_df)

print("\n\n\t TASK 3: Lookup operation and automation")

scores_dict =dict(zip(scores_df['student_id'],scores_df['exam_score']))
print("\n\nscores:\n",scores_dict)
students_df['exam_score'] = students_df['student_id'].map(scores_dict)
print("\n\n\nUpdated Student DataFrame:\n",students_df)

scores_left_join=pd.merge(students_df,scores_df,on='student_id',how='left')
print("\n\n\nComparision BETWEEN TWO :\n",scores_left_join)

def auto_merge(df1, df2, key_column,join_type):
  merged_df=pd.merge(df1,df2,on=key_column,how=join_type)
  return {
    'result_df' : merged_df,
    'row_count' : merged_df.shape[0],
    'type' : join_type
         }

test_left=auto_merge(students_df,scores_df,'student_id','left')
print("\n\n\nTesting Function :\n",test_left)
test_right=auto_merge(students_df,scores_df,'student_id','right')
print("\n\n\nTesting Right Function :\n",test_right)
```
