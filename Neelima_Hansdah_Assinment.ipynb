```python

from locale import normalize

import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

# Calculate the mean temperature for each day
mean_temperatures = np.mean(data, axis=1)
print("Shape of the data array:", data.shape)
print("Mean temperatures for each day:", mean_temperatures)

high_temperatures = data[data > 28.0]
high_temp=high_temperatures.flatten()
print("Indices of temperatures above 28 degrees:", high_temp)

normalized_data = (data - data.min()) / (data.max() - data.min())
print("Normalized data:\n", np.round(normalized_data, 2))
```
