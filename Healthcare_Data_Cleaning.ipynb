```python
import pandas as pd
import numpy as np

data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                   111, 112, 113, 114, 115, 101, 107, 118, 119, 120],
    'age': ['25', '34', None, '45', '29', None, '38', '52', '27', '41',
            '33', 'unknown', '48', '26', '35', '25', '38', '31', None, '44'],
    'weight': ['70', '65', '80', None, '75', None, '68', '90', '72', '85',
               '78', None, '82', '69', 'N/A', '70', '68', '74', None, '88'],
    'blood_pressure': [120, 130, None, 140, 125, None, 135, None, 118, 145,
                      128, None, 138, 122, None, 120, 135, 126, None, 142],
    'medication': ['Aspirin', 'Metformin', 'Lisinopril', None, 'Aspirin',
                   'Metformin', 'Lisinopril', 'Aspirin', None, 'Metformin',
                   'Lisinopril', 'Aspirin', None, 'Metformin', 'Aspirin',
                   'Aspirin', 'Lisinopril', 'Metformin', 'Aspirin', None],
    'insurance_provider': ['Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None,
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', 'Blue Cross',
                          'Aetna', None, 'UnitedHealth', 'Blue Cross', 'Aetna',
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None]
}

df = pd.DataFrame(data)

print("\n\n *** Task1=Inspect the Data *** \n")
df.info()
print("count of missing values: \n", df.isnull().sum())

print("percentage of missing values: \n", (df.isnull().sum() / len(df)) * 100)
initial_duplicates = df.duplicated().sum()
print("count of duplicate values: ", initial_duplicates)

intial_dtypes = df.dtypes
print("\n initial data types:", intial_dtypes)

print("\n\n *** Task2 = Data Type Conversion *** \n")

df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
print("data types of age and weight: ", df['age'].dtype, df['weight'].dtype)
initial_missing = df.isnull().sum().sum()
print("count of missing values after conversion: \n", initial_missing )

df['insurance_provider'] = df['insurance_provider'].astype('category')
print("\n data type of insurance_provider: ", df['insurance_provider'].dtype)

df.dtypes


print("\n\n *** Task3 = Handling Missing Values *** \n")


df['age'] = df['age'].fillna(df['age'].median())
df['weight'] = df['weight'].fillna(df['weight'].median())
df['blood_pressure'] = df['blood_pressure'].fillna(df['blood_pressure'].median())
df['medication'] = df['medication'].fillna(df['medication'].mode()[0])
df['insurance_provider'] = df['insurance_provider'].fillna(df['insurance_provider'].mode()[0])

print("\n count of missing values after imputation: \n", df.isnull().sum())

df[df.duplicated(keep=False)]
df.duplicated(subset=['patient_id'])
df1 = df.drop_duplicates(subset=['patient_id'], keep='first')


print("\n\n *** Task4 = Handling Duplicates *** \n")

print ("\n previous and final data shapes: \n", df.shape, df1.shape)

print("\n\n *** Task5 = Complete Workflow with verification *** \n")

df_clean = df1.copy()
df_clean.info()
df_clean.dtypes
print("\n count of missing values in cleaned data: \n", df_clean.isnull().sum())
df_clean[df_clean.duplicated(keep=False)]
print (f"\n Shape:{df.shape}->{df_clean.shape}")

print (f"\n count of missing values: {initial_missing} -> {df_clean.isnull().sum().sum()}")

print (f"\n count of duplicate values: {initial_duplicates} -> {df_clean.duplicated().sum()}")

print (f"\n Initial data types: \n{intial_dtypes} \n\n Final data types: \n{df_clean.dtypes}")



 



```
